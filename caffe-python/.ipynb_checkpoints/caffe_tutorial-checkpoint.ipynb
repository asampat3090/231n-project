```python
# The blobs expose their data as numpy arrays, so they are easy to work with in
# python. For example we can easily visualize the conv filters of our network:

def vis_square(data, padsize=1, padval=0):
    data -= data.min()
    data /= data.max()
    
    # force the number of filters to be square
    n = int(np.ceil(np.sqrt(data.shape[0])))
    padding = ((0, n ** 2 - data.shape[0]), (0, padsize), (0, padsize)) + ((0, 0),) * (data.ndim - 3)
    data = np.pad(data, padding, mode='constant', constant_values=(padval, padval))
    
    # tile the filters into an image
    data = data.reshape((n, n) + data.shape[1:]).transpose((0, 2, 1, 3) + tuple(range(4, data.ndim + 1)))
    data = data.reshape((n * data.shape[1], n * data.shape[3]) + data.shape[4:])
    plt.axis('off')
    plt.imshow(data)
    plt.gcf().set_size_inches(12, 12)
    plt.show()

conv1_filters = net.params['conv1'][0].data
print conv1_filters.shape
vis_square(conv1_filters.transpose(0, 2, 3, 1),padval=1)
# idx = 0
# for i in xrange(10):
#   for j in xrange(10):
#     if idx >= conv1_filters.shape[0]: break
#     f = conv1_filters[idx].transpose(1, 2, 0)
#     f_min, f_max = np.min(f), np.max(f)
#     f = 255.0 * (f - f_min) / (f_max - f_min)
#     idx += 1
#     plt.subplot(10, 10, idx)
#     plt.imshow(f.astype('uint8'))
#     plt.axis('off')
# plt.gcf().set_size_inches(12, 12)
# plt.show()
```

```python
# We will need access to the ImageNet mean file in order to preprocess our data:

# Caffe ships with a .npy version of the ImageNet mean file. You should use this if
# you fine-tune a network that was pretrained on ImageNet. If you train a network from
# scratch then you should instead compute the mean over your own training data.

# mean_file =  os.path.join(os.environ['CAFFE'], 'python/caffe/imagenet/ilsvrc_2012_mean.npy')
# mean = np.load(mean_file)
# print mean.shape

# Convert mean file
mean_path = '../caffe/models/hybridCNN/hybridCNN_mean.binaryproto'
# mean_path = '../caffe/models/placesCNN/places205CNN_mean.binaryproto'
blob = caffe.proto.caffe_pb2.BlobProto()
data = open(mean_path, 'rb' ).read()
blob.ParseFromString(data)
arr = np.array( caffe.io.blobproto_to_array(blob) )
mean = arr[0]

low, high = np.min(mean), np.max(mean)
mean = 255.0 * (mean - low) / (high - low)
plt.imshow(mean[(2, 1, 0), :, :].transpose(1, 2, 0).astype('uint8'))
plt.axis('off')
plt.show()
```

```python
# VISUALIZE DATASET
datadir = '../caffe/data/screenshotsdb/large_test'

files = os.listdir(datadir)
calendar_f = []
email_f =[]
programming_f = []
social_media_f = []
word_processing_f = []
web_shopping_f = []
finance_f = []
news_f = []
music_editing_f = []
picture_editing_f = []
video_editing_f = []
watching_video_f = []
web_browsing_f = []
web_forum_f = []


for f in files: 
    c = os.path.splitext(f)[0].split('_')[0]
    c2 = os.path.splitext(f)[0].split('_')[1]
    if c == 'calendar': 
        calendar_f.append(f)
    elif c=='email':
        email_f.append(f)
    elif c=='programming':
        programming_f.append(f)
    elif c=='social':
        social_media_f.append(f)
    elif c=='word':
        word_processing_f.append(f)
    elif c=='web' and c2=='shopping':
        web_shopping_f.append(f)
    elif c=='finance':
        finance_f.append(f)
    elif c=='news':
        news_f.append(f)
    elif c=='music':
        music_editing_f.append(f)
    elif c=='picture':
        picture_editing_f.append(f)
    elif c=='video':
        video_editing_f.append(f)
    elif c=='watching':
        watching_video_f.append(f)
    elif c=='web' and c2=='browsing':
        web_browsing_f.append(f)
    else: 
        web_forum_f.append(f)

all_class_list = [calendar_f,email_f,programming_f,social_media_f,word_processing_f,web_shopping_f,
                  finance_f,news_f,music_editing_f,picture_editing_f,video_editing_f,watching_video_f,web_browsing_f,web_forum_f]

# Visualize some examples from all classes 
fig = plt.figure()
examples_per_class = 14
classes = ['calendar','email','programming','social_media','word_processing','web_shopping',
           'finance','news','music_editing','picture_editing','video_editing','watching_video','web_browsing','web_forum']
for cls, cls_name in enumerate(classes):
    random.shuffle(all_class_list[cls])
    subset = all_class_list[cls][:examples_per_class]

    for i,f in enumerate(subset): 
        filepath = os.path.join(datadir,f)
        img = imread(filepath)
        plt.subplot(examples_per_class, len(classes), i * len(classes) + cls + 1)
        plt.imshow(img.astype('uint8'))
        plt.axis('off')
        if i == 0:
            plt.title(str(cls))
plt.autoscale(False)
fig.savefig('filename.eps', format='eps',interpolation='no')
plt.show()
```

```python
# TEST DATASET
data_dir = '/Users/anandsampat/project_231n/caffe/data/screenshotsdb/small_test'
test_accuracy = np.zeros(3)

for i in xrange(3):
    files = os.listdir(data_dir)
    random.shuffle(files)
    num_files = 10
    files = files[:num_files]
    counter = 0
    for f in files:
        if counter % 100 ==0: 
            print "processed 100"
        filepath = os.path.join(data_dir,f)
        img = imread(filepath)
        if img.shape!=(227,227): #transform image if not 256x256
        #     plt.imshow(img)
        #     plt.title('Input image')
        #     plt.axis('off')
        #     plt.show()

            # Before feeding the image to the network, we need to preprocess it:
            # 1) Resize image to (256, 256)
            # 2) Swap channels from RGB to BGR (for CaffeNet)
            # 3) Reshape from (H, W, K) to (K, H, W)
            # 4) Subtract ImageNet mean
            # 5) Crop or resize to (227, 227)

            # Resize image to same size as mean
            H_mean, W_mean = mean.shape[1:]
            img = imresize(img, (H_mean, W_mean))
        #     plt.imshow(img)
        #     plt.title('Resized image')
        #     plt.axis('off')
        #     plt.show()

            # Swap channels from RGB to BGR
            try:
                img = img[:, :, (2, 0, 1)]
            #     plt.imshow(img)
            #     plt.title('Channel-swapped image')
            #     plt.axis('off')
            #     plt.show()

                # Reshape from (H, W, K) to (K, H, W)
                img = img.transpose(2, 0, 1)
            except: 
                num_files = num_files-1
                pass

            # Subtract mean
            img = img - mean

            # Crop to input size of network
            H_in, W_in = net.blobs['data'].data.shape[2:]
            H0 = (H_mean - H_in) / 2
            H1 = H0 + H_in
            W0 = (W_mean - W_in) / 2
            W1 = W0 + W_in
            img = img[:, H0:H1, W0:W1]

        # Copy input data to input blob of the network
        net.blobs['data'].data[0] = img

        # Call net forward
        net.forward()

        # Pull out the probabilities from the network and print them
        probs = net.blobs['prob'].data[0]
        word_arr = os.path.splitext(f)[0].split('_')
        if len(word_arr)>2:
            c = word_arr[0]+'_'+word_arr[1]
        else: 
            c = word_arr[0]
    #     print tasks[probs.argmax()]
        if tasks[probs.argmax()] == c: 
            test_accuracy[i] += 1
            print test_accuracy[i]
        counter += 1
    test_accuracy[i] = float(test_accuracy[i])/num_files
    print test_accuracy[i]
print np.mean(test_accuracy)
```
